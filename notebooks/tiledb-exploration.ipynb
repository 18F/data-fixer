# Exploration of TileDB with sample data

## Retrieve sample data

```python
# Get sample CSV
import urllib.request
urllib.request.urlretrieve('https://covid.ourworldindata.org/data/owid-covid-data.csv', 'owid-covid-data.csv')
```

```python
# Confirm that this is a valid CSV
import pandas
pandas.read_csv('owid-covid-data.csv')
```

## Load data

```python
# If already exists, delete first.
import shutil
shutil.rmtree('owid-covid-data.tldb', ignore_errors=True)

import numpy
import tiledb
column_types = {'iso_code': numpy.str, 'continent': numpy.str, 'location': numpy.str, 'tests_units': numpy.str}
fillna = {column: '' for column in column_types.keys()}
tiledb.from_csv('owid-covid-data.tldb', 'owid-covid-data.csv', fillna=fillna, column_types=column_types)

df = tiledb.open_dataframe('owid-covid-data.tldb')
df
```
